### Import Liblary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Baca File CSV

```python
df = pd.read_csv("Bank Customer Churn Prediction.csv")
```

```python
df.head()
```

```python
data_pie = df['country'].value_counts().rename_axis('Country').reset_index(name='id_count')
plt.figure(figsize=(9, 9))
plt.pie(data_pie["id_count"], labels = data_pie["Country"], startangle = 90, autopct='%.1f%%', explode=(0.02,0,0))
plt.title('Country')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Dari Diagram Pie di atas tampak bahwa nasabah ABC Multistate bank yang berasal dari France(50.1%), Germany 25.1% dan Spain(24.8%)

### Bar PLot

```python
plt.figure(figsize = (10, 5))
Attrplot = sns.countplot(x = 'gender', data = df, palette='flare')
plt.title("Gender")
plt.ylabel('Jumlah')
plt.xlabel('Gender')
plt.grid()
plt.show()
```

Dari data di atas tampak bahwa Nasabah Laki-Laki lebih banyak daripada perempuan

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='country',data=df, hue='gender', palette='flare')
plt.title('Perbandingan Gender Nasabah di Tiap Negara')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('Jumlah')
plt.xlabel('Negara')
plt.grid()
plt.show()
```

Dari data diatas terlihat bahwa nasabah laki-laki di Prancis lebih banyak daripada perempuan, sama halnya di negara Spanyol dan Jerman

```python
sns.boxplot(x='gender',data=df,y='tenure', palette='flare')
plt.title('Data Gender Nasabah terhadap Lama Pengaksesan Bank ABC Multistate ')
plt.ylabel('Pengaksesan Bank (Tahun)')
plt.xlabel('Gender')
```
